### Visualization


```python
plt.figure()
sns.scatterplot(
    data=data,
    x='SepalLengthCm',
    y='SepalWidthCm',
    hue='Species'
)
plt.title('Sepal Length vs Width')
plt.show()
```

## Results and Insights

The scatter plot of sepal length versus sepal width reveals that Iris-setosa forms a distinct and well-separated cluster, indicating that it can be easily classified based on these features. While, Iris-versicolor and Iris-virginica exhibit considerable overlap, making them harder to distinguish using only sepal measurements.

```python
plt.figure()
sns.histplot(data['PetalLengthCm'], bins=20, kde=True)
plt.title('Petal Length Distribution')
plt.show()
```

## Results and Insights

The histogram of petal length reveals a clear two peaks distribution, indicating the presence of two distinct groups within the dataset. The lower range (approximately 1–2 cm) corresponds to Iris-setosa, while the higher range (4–6 cm) represents Iris-versicolor and Iris-virginica. The minimal overlap between these ranges suggests that petal length is a highly effective feature for distinguishing species.


```python
plt.figure()
sns.histplot(data['SepalLengthCm'], bins=20, kde=True)
plt.title('Sepal Length Distribution')
plt.show()
```

## Results 
The sepal length distribution appears more continuous and approximately normal, with values spread across a wider range. Significant overlap is observed among the species, indicating that sepal length alone is not a strong discriminative feature.


```python
plt.figure()
sns.boxplot(
    data=data,
    x='Species',
    y='PetalLengthCm'
)
plt.title('Petal Length by Species')
plt.show()
```

## Results and Insights

The box plot of petal length across species shows a clear separation between the three Iris classes. Iris-setosa exhibits very small petal lengths with minimal variation, making it completely distinguishable from the other species. Iris-versicolor displays moderate petal lengths with some variability, while Iris-virginica has the largest petal lengths and a wider spread of values.

Although a slight overlap exists between Iris-versicolor and Iris-virginica, the overall trend demonstrates a consistent increase in petal length from setosa to virginica. This indicates that petal length is a highly effective feature for species classification, particularly for identifying Iris-setosa.

```python
plt.figure()
sns.boxplot(
    data=data,
    x='Species',
    y='SepalLengthCm'
)
plt.title('Sepal Length by Species')
plt.show()
```

## Results and Insights
The box plot analysis shows a clear trend in sepal length across species. Iris-setosa has the smallest median sepal length with low variability, while Iris-versicolor shows moderate values. Iris-virginica has the highest median sepal length along with greater variability and the presence of potential outliers.

```python
sns.pairplot(data, hue='Species')
plt.show()
```

### Conclusion

The Iris dataset analysis demonstrated how exploratory data analysis and visualization can reveal important patterns in data.
By analyzing the Iris dataset, it became clear that some features are more useful than others for identifying flower species.Visualizations showed that Iris-setosa is clearly distinguishable from the other species, while Iris-versicolor and Iris-virginica have some overlapping characteristics. Among all features, petal length and petal width were the most effective for classification, whereas sepal features were less distinct.
